2.Normalize the data with min-max scaling

```python
import pandas as pd
d=pd.read_csv("/train.csv")
print(d)
```

```python
x=d['price_range']
print(x)
```

```python
y=d.drop('price_range',axis=1)
print(y)
```

```python
df=d-d.min()/d.max()-d.min()
print(df)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(y,x,test_size=0.3)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
